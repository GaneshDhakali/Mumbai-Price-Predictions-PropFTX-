```python
import pandas as pd
```

```python
df = pd.read_excel("Assignment Data Scientist.xlsx")
```

```python
df
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df.head(10)
```

```python
df["City"].value_counts()
```

```python
#Filtered dataset to Mumbai due to schema inconsistencies in Hyderabad rows

df = df[df["City"]== "Mumbai"]
```

```python
df.shape
```

```python
#Columns with higher missing values removed
df = df.drop(columns = ["Growth Type","Unnamed: 8"])
```

```python
df.info()
```

```python
df.head(15)
```

```python
# coverting the average price from text to numeric for modelling
df["Average Price"] = pd.to_numeric(df["Average Price"], errors = "coerce")
```

```python
df.info()
```

```python
df["Average Price"].isnull().sum()
```

```python
# Removing rows where Average Price (target) is NaN
df = df.dropna(subset=["Average Price"])
```

```python
df.shape
```

```python
# Counting how many times each property type appears
df["Type"].value_counts()
```

```python
# Removing non-informative columns with no variation
df = df.drop(columns=["City","Type"])
```

```python
df.columns

```

```python
#Dropping price range as Average price already summarizes it
df = df.drop(columns = ["Price Range"])
```

```python
#Dropping QoQ as it has many missing vaklues
df = df.drop(columns = ["Q-o-Q"])
```

```python
#Defining taget variable
y = df["Average Price"]

# Defining feature set
x = df.drop(columns=["Average Price"])
```

```python
# Encoding categorical features into numeric using One-Hot Encoding
x = pd.get_dummies(x)
```

```python
x.shape
```

```python
#Model Learning
```

```python
from sklearn.model_selection import train_test_split
```

```python
# Splitting dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=42)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
# Choosing Random Forest Regression since it can capture non-linear patterns in property pricing and 
# works well with high-dimensional encoded categorical data
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
# Initializing Random Forest regression model
model = RandomForestRegressor(random_state=42)
```

```python
# Training model on training data
model.fit(X_train, y_train)
```

```python
df.columns

```

```python
# Predicting on test data
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
# Calculating Root Mean Squared Error to evaluate prediction accuracy
import numpy as np

rmse = np.sqrt(mean_squared_error(y_test, y_pred))
rmse
```

```python
# Insight:
# The model achieved an RMSE of approximately 3744, indicating moderate prediction error is present.
# Given the inherent variability in real estate pricing across localities and time periods,
# this performance is acceptable for a baseline model.
```

```python
# Next Saving trained model for API deployment
import pickle

pickle.dump(model, open("price_model.pkl", "wb"))
```

```python
### Distribution of Average Property Prices

##This plot shows how property prices are distributed across the dataset. 
##It helps identify skewness, spread, and presence of premium pricing clusters.
```

```python
# Visualizing distribution of the target variable (Average Price)
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
plt.hist(y, bins=30)
plt.title("Distribution of Average Property Price")
plt.xlabel("Average Price")
plt.ylabel("Frequency")
plt.show()
```

```python
#Insight:
# Property prices shows variability in property prices across localities. 
```

```python
### Feature Importance

#This plot highlights the most influential encoded features affecting price prediction.
```

```python
# Extracting feature importance from Random Forest model
import pandas as pd

importance = pd.Series(model.feature_importances_, index=x.columns)
top_features = importance.sort_values(ascending=False).head(10)

plt.figure(figsize=(10,5))
top_features.plot(kind="bar")
plt.title("Top Important Features in Price Prediction")
plt.show()
```

```python
# Insight:
# Certain locality and temporal features significantly influence price predictions.
```
